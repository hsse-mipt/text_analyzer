```python
import numpy as np
import pandas as pd

basic_data = pd.read_csv('../data/train_data.csv', sep='\t')
columns = ['sentence', 'label']
first_model_data = basic_data[basic_data['label'] != 0][columns]
first_model_data.set_index(np.arange(first_model_data.shape[0]), inplace=True)
first_model_data.replace(to_replace=-1, value=0, inplace=True)
first_model_data
```

<h3>Следующие предложения позитивные:<h3>

```python
positive_sentences = basic_data[basic_data['label'] == 1]
for index, row in positive_sentences.head(5).iterrows():
    print(f"Предложение: {row['sentence']}\nИмя объекта: {row['entity']}\nТип сущности: {row['entity_tag']}\n")
```

<h3>Следующие предложения негативные:<h3>

```python
negative_sentences = basic_data[basic_data['label'] == -1]
for index, row in negative_sentences.head(5).iterrows():
    print(f"Предложение: {row['sentence']}\nИмя объекта: {row['entity']}\nТип сущности: {row['entity_tag']}\n")
```

<h3>Следующие предложения нейтральные:<h3>

```python
neutral_sentences = basic_data[basic_data['label'] == 0]
for index, row in neutral_sentences.head(5).iterrows():
    print(f"Предложение: {row['sentence']}\nИмя объекта: {row['entity']}\nТип сущности: {row['entity_tag']}\n")
```

**Заметки**

-В предложениях встречаются английские слова, которые тоже несут смысл

-Некоторые предложения очень сложные, нужно уметь выявлять причинно-следственные связи, понимать фразеологизмы и тп

Вот, что пишут организаторы:
Особенности  анализа тональности по отношению к сущностям в новостях такие:
- высокая доля нейтральных сущностей, к которым не высказано никакого отношения, т.е. нейтральный класс превалирует,
- в некоторых предложениях новостных текстов наблюдается высокая концентрация именованных сущностей, к которым высказана разная тональность,
т.е. система должна правильно распределить тональности между участниками,
- отношения между разными сущностями несимметричны, например, в предложении "команда X заинтересована в игроке Y" есть позитивная тональность по
отношению к Y, но неизвестна тональность по отношению к X.
